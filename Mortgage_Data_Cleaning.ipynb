```python
import pandas as pd
import matplotlib as plot
import seaborn as sns
```

```python
# Reading excel data
mortgages = pd.read_excel("Table26-2019-by-Month.xls")
mortgages.dropna(inplace=True)
mortgages.rename(columns=mortgages.iloc[0], inplace=True)
mortgages.drop(index=(1), inplace=True)
mortgages.reset_index(inplace=True)
mortgages.drop(columns=['index','Contract Interest Rate (%)'],inplace=True)
mortgages.to_csv("mortgage_rates.csv")
mortgages.head(25)
```

```python
# Visualize over time
sns.lineplot(data=mortgages, x='Year', y='Effective Rate (%)').set_title("Effective Mortgage Rate over Time")
```

```python
# Visualize over time
sns.lineplot(data=mortgages, x='Year', y='Initial Fees and Charges (%)').set_title("Initial Fees and Charges over time")
```

```python
# Visualize over time
sns.lineplot(data=mortgages, x='Year', y='Share of Total Market (%)')
```

```python
# Visualize over time
sns.lineplot(data=mortgages, x='Year', y='Loan-to-Price Ratio (%)').set_title("Loan-to-Price Ratio over time")
```

```python
# Visualize over time
sns.lineplot(data=mortgages, x='Year', y='Purchase Price ($thou)').set_title("Purchase Price over time")
```
